**1).How high was the issuance time of permit in each borough? For which type of job was the issuance time high?**

```python
dob_data['Permit_Filing_Date'] = pd.to_datetime(dob_data['Permit_Filing_Date'],format='%m/%d/%Y %I:%M:%S %p')
#meta = ('Permit_Filing_Date', pd.Timestamp) 
#new_dob_data['Permit_Filing_Date'].map_partitions(pd.to_datetime,meta=meta).compute()
```

```python
dob_data['Permit_Issuance_Date'] = pd.to_datetime(dob_data['Permit_Issuance_Date'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
dob_data.head()
```

```python
dob_data['Issuance_time'] = (dob_data['Permit_Issuance_Date'] - dob_data['Permit_Filing_Date']).dt.days
```

```python
dob_data.head(100)
```

```python
dob_data["Issuance_time"].mean()          #Average Issuance time between filing and issuance of permit
```

```python
#dob_data.groupby("Issuance_time").Borough.value_counts()
dob_data[dob_data['Issuance_time']<0]          #Data Discrepancy
```

```python
dob_data.groupby("Borough").Issuance_time.sum()
```

```python
dob_data.groupby('Borough')['Issuance_time'].mean()         #Finding mean Issuance time per Borough
```

```python
dob_data.groupby('Permit_Type')['Issuance_time'].mean()   #Finding mean Issuance time per Permit Type
```

```python
###We can see from the above that Bronx takes highest number of time to issue a permit. Grilling this down further Alteration permit takes more time to be issued.
```

```python
dob_data.groupby(['Borough','Permit_Type'])['Issuance_time'].mean()
```

```python
#fig, ax = plt.subplots(figsize=(15,7))
dob_data.groupby(['Borough','Permit_Type'])['Issuance_time'].mean().unstack().plot.barh()
plt.title("Average Permit Issuance time in each Borough")
plt.xlabel("Average Permit Issuance Time")
```

**As seen from the above, the average issuance time for a permit for alteration work type is high in all the counties.**

# 2).In which borough are the permits re-issued the most and why?

```python
reissued_dob_data = dob_data.loc[dob_data['Permit_Status']=='RE-ISSUED']
```

```python
reissued_dob_data.head()
```

```python
reissued_dob_data.groupby("Borough").Permit_Status.value_counts().unstack().plot.barh()
plt.xlabel("Number of Permits Re-issued")
```

**As seen in the above graph, Manhattan high number of reissued permits.**

```python
reissued_dob_data.groupby(['Borough'])['Permit_Type'].value_counts().unstack().plot.barh()
plt.title("Boorugh Vs Type of Permit re issued")
plt.xlabel("Number of Re-Issued Permits")
```

**Manhattan has the highest number of re - issued permits. It is the oldest Borough out of all the five, and was developed in 1898. As seen above, Manhattan has the highest EW(Equipment Work) permits re-issued. EW has Boiler servicing, Fuel storage services, Fire alarm servicing, Fire Suppression, mechanical part servicing, other servicing like construction related, pipes and sprinkler serviceing in it.
