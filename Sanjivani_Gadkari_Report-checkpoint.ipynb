```python
import ipywidgets as widgets
from IPython.display import display, HTML

javascript_functions = {False: "hide()", True: "show()"}
button_descriptions  = {False: "Show code", True: "Hide code"}


def toggle_code(state):

    """
    Toggles the JavaScript show()/hide() function on the div.input element.
    """
    output_string = "<script>$(\"div.input\").{}</script>"
    output_args   = (javascript_functions[state],)
    output        = output_string.format(*output_args)

    display(HTML(output))


def button_action(value):

    """
    Calls the toggle_code function and updates the button description.
    """

    state = value.new

    toggle_code(state)

    value.owner.description = button_descriptions[state]


state = False
toggle_code(state)

button = widgets.ToggleButton(state, description = button_descriptions[state])
button.observe(button_action, "value")

display(button)
```

# Analysis on Demolishment of Buildings in Manhattan

**The New York Department of buildings (DOB) is is a department of the New York City government that enforces the city's building codes and zoning regulations, issues building permits, licenses, registers and disciplines certain construction trades, responds to structural emergencies and inspects over 1,000,000 new and existing buildings.**

**The Department of Buildings (DOB) issues permits for construction and demolition activities in the City of New York. The construction industry must submit an application to DOB with details of the construction job they would like to complete. The primary types of application, aka job type, are: New Building, Demolition, and Alterations Type 1, 2, and 3. Each job type can have multiple work types, such as general construction, boiler, elevator, and plumbing. Each work type will receive a separate permit.Each row/record in this dataset represents the life cycle of one permit for one work type. The dataset is updated daily with new records.**

**According to the DOB, every permit is a new project i.e a new repair.When the permit is reissued, means that a repair which was earlier done needs to be done again, hence this is considered as a new repair.**


**This analysis uses the Permit issuance data, complains received data (complains received to DOB regarding buildings),Voilations data (Voilations DOB Violation is a notice that a property is not in compliance with some provision of applicable law) and Population of New York data.**

**This analysis focuses on the buildings which are given a demolition permit by DOB and factors that might have lead to this decision.**

```python
#Converting the Permit filing date from object to datetime format
dob_data['Permit_Filing_Date'] = pd.to_datetime(dob_data['Permit_Filing_Date'],format='%m/%d/%Y %I:%M:%S %p')
#meta = ('Permit_Filing_Date', pd.Timestamp) 
#new_dob_data['Permit_Filing_Date'].map_partitions(pd.to_datetime,meta=meta).compute()
```

```python
#Converting the Permit issuance date from object to datetime format
dob_data['Permit_Issuance_Date'] = pd.to_datetime(dob_data['Permit_Issuance_Date'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
dob_data.head()
```

```python
#Finding the issuance time for each permit
dob_data['Issuance_time'] = (dob_data['Permit_Issuance_Date'] - dob_data['Permit_Filing_Date']).dt.days
```

```python
dob_data.head()
```

```python
#Average Issuance time between filing and issuance of permit
dob_data["Issuance_time"].mean()   
```

**Average time a permit takes to be issued is approximately 4 days.**

**As seen above, in a few rows the issuance time is negative, which would mean data discrepancy. However in most of the rows the Permit status is in process and the filing status is renewable.
